```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('../data/house_pricing.csv')
#data.info()

sb.relplot(x="LotArea",y="SalePrice", data=data)

dataFrame1 = pd.DataFrame(data, columns = ['SalePrice' , 'LotArea']) 
filterData1 = dataFrame1[(data['SalePrice'] < 500000) & (data['LotArea'] < 20000) ]

sb.relplot(x="LotArea",y="SalePrice", data=filterData1)


dataFrame2 = pd.DataFrame(data, columns = ['LotFrontage' , 'LotArea']) 
filterData2 = dataFrame2[(data['LotFrontage'] < 200) & (data['LotArea'] < 100000) ]

sb.relplot(x="LotArea",y="LotFrontage", data=filterData2)

#data[:]
```
